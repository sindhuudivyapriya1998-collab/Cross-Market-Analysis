# **crypto currency**

```python
import requests
import time

crypto_data =[]

for page in range(1,8):
   url = f"https://api.coingecko.com/api/v3/coins/markets?vs_currency=inr&per_page=250&order=market_cap_desc&page={page}&sparkline=False"

   response = requests.get(url)

   if response.status_code == 200:
     data = response.json()
     crypto_data.extend(data)
     time.sleep(20)
```

# **Coin Records**

```python
import requests
import pandas as pd
import time

price_records =[]
for coin_id in ["bitcoin","ethereum","tether"]:
  url = f"https://api.coingecko.com/api/v3/coins/{coin_id}/market_chart?vs_currency=inr&days=365"

  response = requests.get(url)
  if response.status_code == 200:
    coin = response.json()
    for ts, price in coin.get("prices",[]):
      price_records.append({"coin_id":coin_id,"date":ts,"price_inr":price})
price_records
time.sleep(10)
```

```python
len(price_records)
```

```python
import pandas as pd
df_coin = pd.DataFrame(price_records)
df_coin
```

#**Stocks Records**

```python
import yfinance as yf
import pandas as pd

tickers = ["^GSPC","^IXIC","^NSEI"]
start_date = "2020-01-01"
end_date = "2026-03-09"

stocks = yf.download(tickers, start = start_date, end= end_date, group_by="tickers")
stocks.head()
```

```python
df_list =[]
for ticker in tickers:
   text=stocks[ticker].reset_index()
   text['Ticker'] = ticker
   df_list.append(text)
df_all = pd.concat(df_list,ignore_index=True)
df_all
```

```python
df_stocks1 = df_all[df_all['Ticker'] == '^NSEI'].reset_index(drop=True)
df_stocks1
```

```python
df_stocks2 = df_all[df_all['Ticker'] == '^GSPC'].reset_index(drop=True)
df_stocks2
```

```python
df_stocks3 = df_all[df_all['Ticker'] == '^IXIC'].reset_index(drop=True)
df_stocks3
```

```python
df_stocks = pd.concat([df_stocks1,df_stocks2,df_stocks3])
df_stocks
```

```python
df_stocks.to_csv("stocks_data.csv")
```

```python
%%writefile app.py

import streamlit as st
import sqlite3
import pandas as pd

connection = sqlite3.connect("crypto.db")
cursor = connection.cursor()

from streamlit_option_menu import option_menu

with st.sidebar:
    selected = option_menu("CONTENTS", ["📋Cross Market Overview", '📝SQL Query','💰Top 3 Crypto Analysis'], menu_icon="cast", default_index=1)
if selected  == "📋Cross Market Overview":
        st.title("**📊Cross-Market Analysis**")
        st.divider()
        st.caption("crytpto oil stock Market powered analytics")

        col1 , col2 =st.columns(2)
        with col1:
          start_date = st.date_input("start Date")
        with col2:
          end_date = st.date_input("End Date")

        bitcoin_avg = pd.read_sql("select avg(price_inr) as Bitcoin_avg from coin_record where coin_id = 'bitcoin' and date between ? and ?",connection,params =[str(start_date),str(end_date)])
        value = bitcoin_avg.iloc[0,0]
        oil_avg = pd.read_sql("select avg(price_inr) as oil_avg from oil_record where date between ? and ?",connection,params =[str(start_date),str(end_date)])
        value1 = oil_avg.iloc[0,0]
        sp500_avg = pd.read_sql("select avg(close) as sp500_avg from stocks_record where Ticker = '^GSPC' and date between ? and ?",connection,params =[str(start_date),str(end_date)])
        value2 = sp500_avg.iloc[0,0]
        nifty_avg = pd.read_sql("select avg(close) as nifty_avg from stocks_record where Ticker = '^NSEI' and date between ? and ?",connection,params =[str(start_date),str(end_date)])
        value3 = nifty_avg.iloc[0,0]

        col1, col2, col3, col4 = st.columns(4)
        col1.metric("Bitcoin Avg($)",value)
        col2.metric("Oil Avg($)",value1)
        col3.metric("SP500 Avg($)",value2)
        col4.metric("Nifty Avg($)",value3)

        st.header("📈**Daily Market Snapshot**")

        start_date =str(start_date)
        end_date = str(end_date)

        bitcoin = pd.read_sql("""SELECT date(date)as date,price_inr AS bitcoin_price from coin_record WHERE coin_id='bitcoin'and date BETWEEN ? and ?""",connection,params=(start_date,end_date))
        oil = pd.read_sql("""select date(date) as date,price_inr as oil_price from oil_record where date between ? and ?""",connection,params=(start_date,end_date))
        sp500 = pd.read_sql("""select date(Date) as date,close as sp500 from stocks_record where Ticker ='^GSPC'AND date(Date) between ? and ?""",connection,params=(start_date,end_date))
        nifty = pd.read_sql("""select date(Date) as date,close as nifty from stocks_record where Ticker='^NSEI'and date(Date) between ? and ?""",connection,params=(start_date,end_date))

        snapshot = bitcoin.merge(oil, on="date")
        snapshot = snapshot.merge(sp500, on="date")
        snapshot = snapshot.merge(nifty, on="date")

        st.dataframe(snapshot)

if selected  == "📝SQL Query":
      option = st.selectbox(
    "Choose a query:",
    ("1.Find the top 3 cryptocurrencies by market cap.",
     "2.List all coins where circulating supply exceeds 90% of total supply.",
     "3.Get coins that are within 10% of their all-time-high (ATH).",
     "4.Find the average market cap rank of coins with volume above $1B.",
     "5.Get the most recently updated coin.",
     "6.Find the highest daily price of Bitcoin in the last 365 days.",
    "7.Calculate the average daily price of Ethereum in the past 1 year.",
    "8.Show the daily price trend of Bitcoin in January 2025.",
    "9.Find the coin with the highest average price over 1 year.",
    "10.Get the % change in Bitcoin’s price between Sep 2024 and Sep 2025.",
    "11.Find the highest oil price in the last 5 years.",
    "12.Get the average oil price per year.",
    "13.Show oil prices during COVID crash (March–April 2020).",
    "14.Find the lowest price of oil in the last 10 years.",
    "15.Calculate the volatility of oil prices (max-min difference per year).",
    "16.Get all stock prices for a given ticker.",
    "17.Find the highest closing price for NASDAQ (^IXIC).",
    "18.List top 5 days with highest price difference (high - low) for S&P 500 (^GSPC).",
    "19.Get monthly average closing price for each ticker.",
    "20.Get average trading volume of NSEI in 2024.",
    "21.Compare Bitcoin vs Oil average price in 2025.",
    "22.Check if Bitcoin moves with S&P 500 (correlation idea).",
    "23.Compare Ethereum and NASDAQ daily prices for 2025.",
    "24.Find days when oil price spiked and compare with Bitcoin price change.",
    "25.Compare top 3 coins daily price trend vs Nifty (^NSEI).",
    "26.Compare stock prices (^GSPC) with crude oil prices on the same dates.",
    "27.Correlate Bitcoin closing price with crude oil closing price (same date).",
    "28.Compare NASDAQ (^IXIC) with Ethereum price trends.",
    "29.Join top 3 crypto coins with stock indices for 2025.",
    "30.Multi-join: stock prices, oil prices, and Bitcoin prices for daily comparison."))

      if st.button("Run Query"):

        if option == "1.Find the top 3 cryptocurrencies by market cap.":
          df = pd.read_sql("SELECT * FROM crypto_record ORDER BY market_cap DESC limit 3",connection)
          st.write(df)

        if option == "2.List all coins where circulating supply exceeds 90% of total supply.":
          df = pd.read_sql("SELECT * FROM crypto_record WHERE circulating_supply > 0.9 * total_supply",connection )
          st.write(df)

        if option == "3.Get coins that are within 10% of their all-time-high (ATH).":
          df = pd.read_sql("select * FROM crypto_record WHERE current_price >= ath * 0.90",connection)
          st.write(df)

        if option == "4.Find the average market cap rank of coins with volume above $1B.":
          df = pd.read_sql("SELECT avg(market_cap_rank) from crypto_record  WHERE total_volume > 1000000000 ", connection)
          st.write(df)

        if option == "5.Get the most recently updated coin.":
          df = pd.read_sql("SELECT * FROM crypto_record ORDER BY last_updated DESC LIMIT 1", connection)
          st.write(df)

        if option == "6.Find the highest daily price of Bitcoin in the last 365 days.":
          df = pd.read_sql("SELECT *,MAX(price_inr) FROM coin_record where coin_id='bitcoin'",connection)
          st.write(df)

        if option == "7.Calculate the average daily price of Ethereum in the past 1 year.":
          df = pd.read_sql("SELECT *, avg(price_inr) FROM coin_record where coin_id = 'ethereum'",connection)
          st.write(df)

        if option == "8.Show the daily price trend of Bitcoin in January 2025.":
          df = pd.read_sql("SELECT * from coin_record where coin_id='bitcoin' and date between '2026-04-01' and '2026-04-30'",connection)
          st.write(df)

        if option == "9.Find the coin with the highest average price over 1 year.":
          df = pd.read_sql("select coin_id, avg(price_inr) AS avg_price from coin_record GROUP BY coin_id ORDER BY avg(price_inr)DESC LIMIT 1",connection)
          st.write(df)

        if option == "10.Get the % change in Bitcoin’s price between Sep 2024 and Sep 2025.":
          df = pd.read_sql("select ROUND(((select price_inr FROM coin_record WHERE coin_id='bitcoin' ORDER BY date DESC LIMIT 1) - (select price_inr FROM coin_record WHERE coin_id='bitcoin' ORDER BY date ASC LIMIT 1))*100.0 /(select price_inr FROM coin_record WHERE coin_id='bitcoin' ORDER BY date ASC LIMIT 1)) AS percentage_change",connection)
          st.write(df)

        if option == "11.Find the highest oil price in the last 5 years.":
          df = pd.read_sql("select *,max(price_inr) as highest_price from oil_record where date BETWEEN '2021-06-10' AND '2026-06-10'",connection)
          st.write(df)

        if option == "12.Get the average oil price per year.":
          df = pd.read_sql("select strftime('%Y',date) AS Year,Avg(price_inr) AS Avg_price from oil_record group by strftime('%Y',date) order by Year,Avg_price",connection)
          st.write(df)

        if option == "13.Show oil prices during COVID crash (March–April 2020).":
          df = pd.read_sql("select date,price_inr from oil_record where date BETWEEN '2020-03-01' AND '2020-04-30' order by date",connection)
          st.write(df)

        if option == "14.Find the lowest price of oil in the last 10 years.":
          df = pd.read_sql("SELECT *,min(price_inr) as lowest_price FROM oil_record WHERE date BETWEEN '2016-06-10' AND '2026-06-10'",connection)
          st.write(df)

        if option == "15.Calculate the volatility of oil prices (max-min difference per year).":
          df = pd.read_sql("SELECT strftime('%Y',date) AS year, MAX(price_inr) as max_price, min(price_inr) as min_price, ROUND(MAX(price_inr)-min(price_inr),2) AS Volatility FROM oil_record GROUP BY strftime('%Y',date) ORDER BY year",connection)
          st.write(df)

        if option == "16.Get all stock prices for a given ticker.":
          df = pd.read_sql("select close,high,low,Ticker from stocks_record order by Ticker",connection)
          st.write(df)

        if option == "17.Find the highest closing price for NASDAQ (^IXIC).":
          df = pd.read_sql("select max(close) from stocks_record where Ticker == '^IXIC'",connection)
          st.write(df)

        if option == "18.List top 5 days with highest price difference (high - low) for S&P 500 (^GSPC).":
          df = pd.read_sql("select date,high,low,(High-Low) as price_difference from stocks_record WHERE Ticker ='^GSPC' ORDER BY price_difference DESC LIMIT 5",connection)
          st.write(df)

        if option == "19.Get monthly average closing price for each ticker.":
          df = pd.read_sql("select Ticker,strftime('%Y-%m',Date) AS Month, AVG(close) AS avg_close from stocks_record group by Ticker,strftime('%Y-%m',Date) Order by Ticker,Month",connection)
          st.write(df)

        if option == "20.Get average trading volume of NSEI in 2024.":
          df = pd.read_sql("""select avg(volume) as Avg_NSEI from stocks_record where Ticker ='^NSEI' and Date between '2024-01-01' and '2024-12-31'""",connection)
          st.write(df)

        if option == "21.Compare Bitcoin vs Oil average price in 2025.":
          df = pd.read_sql("SELECT(select Avg(price_inr) from coin_record WHERE strftime('%Y',date)='2025') AS Bitcoin_avg_price,(SELECT Avg(price_inr) from oil_record where strftime('%Y',date)='2025') AS Oil_avg_price",connection)
          st.write(df)

        if option == "22.Check if Bitcoin moves with S&P 500 (correlation idea).":
          df = pd.read_sql("SELECT date(coin_record.date) AS Date,coin_record.price_inr AS Bitcoin_price,stocks_record.close AS GSPC_price FROM coin_record JOIN stocks_record ON date(coin_record.date) = date(stocks_record.date) WHERE coin_record.coin_id = 'bitcoin' AND stocks_record.Ticker ='^GSPC'",connection)
          st.write(df)

        if option == "23.Compare Ethereum and NASDAQ daily prices for 2025.":
          df = pd.read_sql("SELECT(select Avg(price_inr) from coin_record WHERE coin_id ='ethereum' AND strftime('%Y',date)='2025') AS Bitcoin_avg_price,(SELECT Avg(price_inr) from oil_record where strftime('%Y',date)='2025') AS Oil_avg_price",connection)
          st.write(df)

        if option == "24.Find days when oil price spiked and compare with Bitcoin price change.":
          df = pd.read_sql("select oil_record.date,oil_record.price_inr AS oil_price, coin_record.price_inr AS Bitcoin_price FROM oil_record JOIN coin_record ON date(oil_record.date) = date(coin_record.date) WHERE coin_record.coin_id = 'bitcoin' AND oil_record.price_inr > (SELECT AVG(price_inr) * 1.10 FROM oil_record) ORDER BY oil_record.date", connection)
          st.write(df)

        if option == "25.Compare top 3 coins daily price trend vs Nifty (^NSEI).":
          df = pd.read_sql("SELECT date(coin_record.date) AS Date, coin_record.coin_id AS coin_name,stocks_record.close AS NSEI_price FROM coin_record JOIN stocks_record ON date(coin_record.date) = date(stocks_record.date) WHERE stocks_record.Ticker = '^NSEI' ORDER BY date, coin_id",connection)
          st.write(df)

        if option == "26.Compare stock prices (^GSPC) with crude oil prices.":
          df = pd.read_sql("""SELECT oil_record.date AS Date,stocks_record.close AS "^GSPC_Stock_price" ,oil_record.price_inr AS oil_price FROM stocks_record JOIN oil_record ON date(stocks_record.date) = date(oil_record.date) WHERE Ticker ='^GSPC' AND date(stocks_record.date) = '2025-01-02'""", connection)
          st.write(df)

        if option == "27.Correlate Bitcoin closing price with crude oil closing price.":
          df =pd.read_sql("""SELECT oil_record.date AS Date,coin_record.price_inr AS Bitcoin_price,oil_record.price_inr AS oil_price FROM coin_record JOIN oil_record ON date(coin_record.date) = date(oil_record.date) WHERE coin_id ='bitcoin' AND date(oil_record.date) ='2025-07-25'""", connection)
          st.write(df)

        if option == "28.Compare NASDAQ (^IXIC) with Ethereum price trends.":
          df= pd.read_sql("""SELECT stocks_record.date,stocks_record.close AS "^IXIC_Stock_price" ,coin_record.price_inr AS Ethereum_price FROM stocks_record JOIN coin_record ON date(stocks_record.date) = date(coin_record.date) WHERE Ticker ='^IXIC' ORDER BY stocks_record.date""", connection)
          st.write(df)

        if option == "29.Join top 3 crypto coins with stock indices for 2025.":
          df =pd.read_sql("""SELECT stocks_record.date as Date,stocks_record.Ticker,stocks_record.close AS stock_price,coin_record.coin_id,coin_record.price_inr AS Coin_price FROM stocks_record JOIN coin_record ON date(stocks_record.Date) = date(coin_record.Date) WHERE strftime('%Y',coin_record.date)='2025' ORDER BY date, coin_id,Ticker""",connection)
          st.write(df)

        if option == "30.Multi-join: stock prices, oil prices, and Bitcoin prices for daily comparison.":
          df=pd.read_sql("""SELECT stocks_record.date,stocks_record.Ticker,stocks_record.close AS Stock_price,coin_record.price_inr AS Bitcoin_price,oil_record.price_inr AS Oil_price FROM stocks_record JOIN oil_record on date(stocks_record.date) = date(oil_record.date) JOIN coin_record on date(stocks_record.date) = date(coin_record.date) WHERE coin_record.coin_id ='bitcoin' ORDER BY stocks_record.date, stocks_record.Ticker""", connection)
          st.write(df)

if selected  == "💰Top 3 Crypto Analysis":
        st.title("🪙Top 3 Crypto Analysis")
        top3 = pd.read_sql("select coin_id from coin_record",connection)
        coin = st.selectbox("Select Cryptocurrency",["Bitcoin","Ethereum","Tether"])

        start_date =st.date_input("Start Date")
        end_date =st.date_input("End Date")

        if coin == "Bitcoin":
          df = pd.read_sql("select date, price_inr FROM coin_record WHERE coin_id ='bitcoin' AND date between ? and ? order by date",connection,params=(start_date,end_date))

        if coin =="Ethereum":
          df = pd.read_sql("select date, price_inr FROM coin_record WHERE coin_id ='ethereum' AND date between ? and ? order by date",connection,params=(start_date,end_date))

        if coin =="Tether":
          df = pd.read_sql("select date, price_inr FROM coin_record WHERE coin_id ='tether' AND date between ? and ? order by date",connection,params=(start_date,end_date))


        st.subheader("📈Daily Price Table")
        st.dataframe(df,use_container_width=True)


```
